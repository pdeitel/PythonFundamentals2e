### Checkpoint 1 Snippets

**(IPython Session)** Use NumPy random-number generation to create an array of six random integers that represent summertime Fahrenheit temperatures in the range 60–100, then perform the following tasks:

a) Convert the array to a Series named `temperatures` and display it.

b) Determine the lowest, highest and average temperatures.

c) Produce descriptive statistics for the Series.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
rng = np.random.default_rng()
```

```python
temps = rng.integers(60, 101, size=6)
```

```python
temperatures = pd.Series(temps)  # (a)
```

```python
temperatures  # (a)
```

```python
temperatures.min()  # (b)
```

```python
temperatures.max()  # (b)
```

```python
temperatures.mean()  # (b)
```

```python
temperatures.describe()  # (c)
```
