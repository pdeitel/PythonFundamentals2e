# 4.11 Arbitrary Argument Lists Self Check

### Checkpoint 1 Snippets

**(Fill-In)** To define a function with an arbitrary argument list, specify a parameter of the form `_____`.

**Answer:** `*args`. (The name `args` is used by convention.)

### Checkpoint 2 Snippets

**(IPython Session)** Create a function named `product` that receives at least one argument and returns the product of all its arguments. Call the function with 10, 20 and 30, then with the sequence of integers produced by `range(1, 6, 2)`.

```python
def product(value1, *args):
    product = value1
    for value in args:
        product *= value
    return product
```

```python
product(10, 20, 30)
```

```python
product(*range(1, 6, 2))  # equivalent to: product(1, 3, 5)
```
