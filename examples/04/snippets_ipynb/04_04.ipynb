# 4.4 Random-Number Generation

### Rolling a Six-Sided Die

```python
import random
```

```python
for roll in range(10):
    print(random.randint(1, 6), end=' ')
```

```python
for roll in range(10):
    print(random.randint(1, 6), end=' ')
```

### Seeding the Random-Number Generator for Reproducibility

```python
random.seed(32)
```
