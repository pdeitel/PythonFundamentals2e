### Checkpoint 1 Snippets

**(True/False)** A function's body is called its suite.

**Answer:** False. Control statements have suites—functions have blocks.

### Checkpoint 2 Snippets

**(True/False)** A function's local variables exist after the function returns to its caller.

**Answer:** False. A function's local variables exist only while it's executing.
