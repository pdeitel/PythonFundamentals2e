### Checkpoint 2 Snippets

**(IPython Session)** Create the `datetime` object `birth` using July 20, 1969, at 10:56 PM (22:56). Write code to display in a locale-specific format the birth date using the `%x` format specifier (which shows only the date) and the birth time using the `%X` format specifier (which shows only the time).

```python
from datetime import datetime
```

```python
birth = datetime.fromisoformat(f'1969-07-20 22:56')
```

```python
print(f'Birth date: {birth:%x}')
```

```python
print(f'Birth time: {birth:%X}')
```

### Checkpoint 3 Snippets

**(IPython Session)** Create a `datetime` object for a future date and calculate the number of days until that date.

```python
future_date = datetime.fromisoformat('2030-01-01')
```

```python
today = datetime.now()
```

```python
difference = future_date - today
```

```python
difference.days
```
