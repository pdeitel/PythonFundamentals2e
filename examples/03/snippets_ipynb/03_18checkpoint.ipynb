# 3.18 Intro to Data Science: Measures of Central Tendency—Mean, Median and Mode Self Check

### Checkpoint 1 Snippets

**(Fill-In)** The `_____`, `_____` and `_____` statistics indicate the average value, most frequently occurring value and the middle value in a set of values, respectively.

**Answer:** mean, mode, median.

### Checkpoint 2 Snippets

**(IPython Session)** For the values 47, 95, 88, 73, 88 and 84, use the `statistics` module to calculate the mean, median and mode.

```python
import statistics
```

```python
values = [47, 95, 88, 73, 88, 84]
```

```python
statistics.mean(values)
```

```python
statistics.median(values)
```

```python
statistics.mode(values)
```

### Checkpoint 3 Snippets

**(IPython Session)** For the values 1, 2, 3, 3 and 1, use the `statistics` module to calculate the mode.

**Answer:** values2 is multimodal because all values have the same frequency.

```python
values2 = [1, 2, 3, 3, 1]
```

```python
statistics.mode(values2)
```
