### Checkpoint 3 Snippets

**(Code)** Write a script that reads an arbitrary number of integer Fahrenheit temperatures from the user and calculates their floating-point average. Assume the user enters temperatures in the range –212 to 212. Use the sentinel value 999. Display the average.

```python
# AverageTemperature.py — Fahrenheit average calculator with sentinel value 999
print('Enter Fahrenheit temperatures one at a time.')

total = 0
count = 0

temp = int(input('Enter -212 to 212 (999 to quit): '))

while temp != 999:
    total += temp
    count += 1
    temp = int(input('Enter -212 to 212 (999 to quit): '))

if count > 0:
    average = total / count
    print(f'\nAverage temperature: {average:.2f}')
else:
    print('\nNo temperatures entered.')
```
