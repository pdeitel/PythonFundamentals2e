### Compound-Interest Problem Requirements Statement

### Calculating Compound Interest

```python
principal = Decimal('1000.00')
```

```python
rate = Decimal('0.05')
```

```python
for year in range(1, 11):
    amount = principal * (1 + rate) ** year
    print(f'{year:>2}{amount:>10.2f}   {amount}')
```
