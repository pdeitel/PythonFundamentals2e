### Checkpoint 2 Snippets

**(Pseudocode)** Write pseudocode for a multiple-selection statement that displays whether the value of a `count` is greater than 10, equal to 10 or less than 10.

**Answer:**
```
If count is greater than 10
    Print "count is greater than 10"
Else If count is equal to 10
    Print "count is equal to 10"
Else
    Print "count is less than 10"
```

### Checkpoint 3 Snippets

**(Code)** Write an `if…elif…else` statement for the pseudocode in Checkpoint 2. Assume the variable `count` exists and contains an int.

````python
if count > 10:
    print('count is greater than 10')
elif count == 10:
    print('count is equal to 10')
else:
    print('count is less than 10')
````
