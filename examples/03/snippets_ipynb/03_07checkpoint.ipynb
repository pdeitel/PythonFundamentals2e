### Checkpoint 2 Snippets

**(IPython Session)** Create a `match…case` statement that displays a letter grade for a corresponding numeric grade point average. Use a four-point scale in which 4 is A, 3 is B, 2 is C, 1 is D and 0 is F.

```python
gpa = 3
```

```python
match gpa:
    case 4:
        print('A')
    case 3:
        print('B')
    case 2:
        print('C')
    case 1:
        print('D')
    case 0:
        print('F')
    case _:
        print('Invalid GPA')
```
