### 6.3 Checkpoint Snippets

**(True/False)** Sets are collections of unique mutable or immutable objects.

**Answer:** False. Sets are collections of unique, immutable objects.

**(Fill-In)** You can create a set from an iterable object containing values by using the built-in `_____` function.

**Answer:** `set`.

**(IPython Session)** Assign the following string to the variable `text`, split it into tokens using the str method `split`, and create a set from the results. Display the unique words in sorted order.
```
'to be or not to be that is the question'
```

````python
text = 'to be or not to be that is the question'
````

````python
unique_words = set(text.split())
````

````python
for word in sorted(unique_words):
    print(word, end=' ')
````

### 6.3.2 Checkpoint Snippets

**(Fill-In)** Two sets are `_____` if they have no elements in common.

**Answer:** disjoint.

**(IPython Session)** Given the sets `{10, 20, 30}` and `{5, 10, 15, 20}`, use the mathematical set operators to produce the following sets:

a) `{30}`

b) `{5, 15, 30}`

c) `{5, 10, 15, 20, 30}`

d) `{10, 20}`

````python
{10, 20, 30} - {5, 10, 15, 20}
````

````python
{10, 20, 30} ^ {5, 10, 15, 20}
````

````python
{10, 20, 30} | {5, 10, 15, 20}
````

````python
{10, 20, 30} & {5, 10, 15, 20}
````
