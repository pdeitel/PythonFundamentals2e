### 6.2.2 Checkpoint Snippets

**(Fill-In)** The dictionary method `_____` returns each key–value pair as a tuple.

**Answer:** `items`.

**(IPython Session)** Given the following dictionary of names and lists of grades, iterate through its key–value pairs and display each student's name and average grade:
```
grades = {'Sue': [98, 87, 94], 'Bob': [84, 95, 91]}
```

````python
grades = {'Sue': [98, 87, 94], 'Bob': [84, 95, 91]}
````

````python
for name, grade_list in grades.items():
    average = sum(grade_list) / len(grade_list)
    print(f"{name}'s average grade is {average:.2f}")
````

### 6.2.3 Checkpoint Snippets

**(Fill-In)** What does an expression of the form `dictionaryName[key] = value` do when the key is not in the dictionary?

**Answer:** It creates the key–value pair for the specified key and value.

**(Fill-In)** What does the same expression do when the key is in the dictionary?

**Answer:** It updates the value associated with the key, replacing the original value.

### 6.2.4 Checkpoint Snippets

**(Fill-In)** The dictionary method `_____` returns a view of the dictionary's keys.

**Answer:** `keys`.

**(True/False)** A view has its own copy of the corresponding data from the dictionary.

**Answer:** False. A view does not have its own copy of the corresponding data from the dictionary. When you iterate over a view, it accesses the dictionary's current contents.

**(IPython Session)** Use a comprehension to create a list of 50 random integers in the range 1–5. Summarize them with a `Counter`. Display the results in a two-column format.

````python
import random
````

````python
numbers = [random.randint(1, 5) for i in range(50)]
````

````python
from collections import Counter
````

````python
counter = Counter(numbers)
````

````python
for value, count in sorted(counter.items()):
    print(f'{value:<4}{count}')
````
