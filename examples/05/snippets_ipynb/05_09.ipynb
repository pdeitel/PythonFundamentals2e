# 5.9 Searching Sequences

### List Method index

```python
numbers = [3, 7, 1, 4, 2, 8, 5, 6]
```

```python
numbers.index(5)
```

### Specifying the Starting Index of a Search

```python
numbers *= 2
```

```python
numbers
```

```python
numbers.index(5, 7)
```

### Specifying the Starting and Ending Indices of a Search

```python
numbers.index(7, 0, 4)
```

### Operators in and not in

```python
1000 in numbers
```

```python
5 in numbers
```

```python
1000 not in numbers
```

```python
5 not in numbers
```

### Using Operator in to Prevent a ValueError

```python
key = 1000
```

```python
position = numbers.index(key) if key in numbers else None
```

```python
if position is not None:
    print(f'found {key} at index {position}')
else:
    print(f'{key} not found')
```

### Built-In Functions any and all

```python
numbers = [0, 7, 2, 0, 11]
```

```python
any(numbers)  # at least one element in numbers is True
```

```python
all(numbers)  # all elements in numbers are True
```

```python
numbers = [3, 7, 2, 5, 11]
```

```python
any(numbers)
```

```python
all(numbers)
```
