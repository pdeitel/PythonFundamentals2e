**(IPython Session)** Create a five-element list containing 67, 12, 46, 43 and 13, then use the list method `index` to search for 43 and 44. Ensure that no `ValueError` occurs when searching for 44.

```python
numbers = [67, 12, 46, 43, 13]
```

```python
numbers.index(43)
```

```python
if 44 in numbers:
    print(f'Found 44 at index: {numbers.index(44)}')
else:
    print('44 not found')
```
