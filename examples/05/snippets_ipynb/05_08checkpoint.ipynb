### Checkpoint 2 Snippets

**(True/False)** All sequences provide a `sort` method.

**Answer:** False. Immutable tuples and strings do not provide a `sort` method. However, you can sort any sequence without modifying it by using the built-in `sorted` function, which returns a new list containing the sorted elements of its argument sequence.
