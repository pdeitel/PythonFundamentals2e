### Checkpoint 1 Snippets

**(Fill-In)** A generator expression uses a technique called `_____`—it produces values on demand.

**Answer:** lazy evaluation.

### Checkpoint 2 Snippets

**(IPython Session)** Create a generator expression that cubes the even integers in a list containing 10, 3, 7, 1, 9, 4 and 2. Use the `list` function to create a list of the results.

```python
numbers = [10, 3, 7, 1, 9, 4, 2]
```

```python
list(x ** 3 for x in numbers if x % 2 == 0)
```

### Checkpoint 3 Snippets

**(IPython Session)** Create a generator function that receives an iterable object containing numbers and, for each value, returns a tuple containing the original value and its cube. Use the `list` function to create a list of the results.

```python
def value_and_cube(values):
    """Returns each value and its cube on demand as a tuple"""
    for value in values:
        yield (value, value ** 3)
```

```python
numbers = [1, 2, 3, 4, 5]
```

```python
list(value_and_cube(numbers))
```
