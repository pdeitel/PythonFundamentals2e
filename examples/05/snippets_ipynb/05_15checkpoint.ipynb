### Checkpoint 2 Snippets

**(IPython Session)** Create the list `foods` containing `'Cookies'`, `'pizza'`, `'Grapes'`, `'apples'`, `'steak'` and `'Bacon'`. Find the smallest string lexicographically then alphabetically with `min`.

```python
foods = ['Cookies', 'pizza', 'Grapes',
         'apples', 'steak', 'Bacon']
```

```python
min(foods)
```

```python
min(foods, key=lambda s: s.lower())
```

**(IPython Session)** Use `zip` with `[10, 20, 30, 40]` and `[1, 2, 3, 4]` to create a new list containing the sums of the elements at the same index.

```python
[(a + b) for a, b in zip([10, 20, 30, 40], [1, 2, 3, 4])]
```
