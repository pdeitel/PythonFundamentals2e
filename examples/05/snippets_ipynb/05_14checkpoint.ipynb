### Checkpoint 2 Snippets

**(IPython Session)** Use the `functools` module's `reduce` function to implement a custom reduction that uses a `lambda` to sum the squares of the values in the list `[10, 20, 30, 40, 50]`. The result should be 5500.

```python
from functools import reduce
```

```python
numbers = [10, 20, 30, 40, 50]
```

```python
reduce(lambda x, y: x + y * y, numbers, 0)
```

### Checkpoint 3 Snippets

**(IPython Session)** Create custom reductions that use lambdas to locate the smallest and largest values in the list `[5, 2, 1, 4, 3]`.

```python
numbers2 = [5, 2, 1, 4, 3]
```

```python
reduce(lambda x, y: x if x <= y else y, numbers2)  # minimum
```

```python
reduce(lambda x, y: x if x >= y else y, numbers2)  # maximum
```
