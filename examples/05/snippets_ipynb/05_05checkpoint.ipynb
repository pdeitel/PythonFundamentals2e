# 5.5 Sequence Slicing Self Check

### Checkpoint 1 Snippets

**(True/False)** Slice operations that modify a sequence work identically for lists, tuples and strings.

**Answer:** False. Only those that do not modify a sequence work identically. Operations that modify sequences cause TypeErrors when used on immutable tuples and strings.

### Checkpoint 2 Snippets

**(IPython Session)** Create a string called `letters` containing the alphabet, then use slices to perform the following operations:

a) Select the first three letters.

b) Select the last three letters.

c) Select every second letter starting from index 0.

d) Select the string with its characters reversed.

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
```

### Checkpoint 3 Snippets

**(IPython Session)** Create a list called `numbers` containing the values 1–15, then use slices to perform the following operations:

a) Select the even integers.

b) Replace the elements at indices 5 through 9 with 0s. Show the modified list.

c) Keep only the first five elements. Show the modified list.

d) Delete all the remaining elements by assigning to a slice. Show the modified list.

```python
numbers = list(range(1, 16))
```

```python
numbers
```

```python
numbers[1::2]  # or numbers[1:len(numbers):2]
```

```python
numbers[5:10] = [0] * len(numbers[5:10])
```

```python
numbers
```

```python
numbers[5:] = []
```

```python
numbers
```

```python
numbers[:] = []
```

```python
numbers
```
