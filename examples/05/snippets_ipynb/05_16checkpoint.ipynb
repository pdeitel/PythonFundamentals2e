### Checkpoint 2 Snippets

**(Two-Dimensional Array)** Consider a two-by-three integer list `t`.

a) How many rows does `t` have? **2**

b) How many columns does `t` have? **3**

c) How many elements does `t` have? **6**

d) What are the names of the elements in row 1? `t[1][0], t[1][1], t[1][2]`

e) What are the names of the elements in column 2? `t[0][2], t[1][2]`

f) Write a nested for statement that sets each element to the sum of its indices.

````python
for row in range(len(t)):
    for column in range(len(t[row])):
        t[row][column] = row + column
````

### Checkpoint 3 Snippets

**(IPython Session)** Given the two-by-three integer list `t = [[10, 7, 3], [20, 4, 17]]`:

a) Determine and display the average of `t`'s elements using nested `for` statements.

b) Use a `for` statement that determines the average using the reductions `sum` and `len`.

````python
t = [[10, 7, 3], [20, 4, 17]]
````

````python
total = 0
````

````python
items = 0
````

````python
for row in t:
    for item in row:
        total += item
        items += 1
````

````python
total / items
````

````python
total = 0
````

````python
items = 0
````

````python
for row in t:
    total += sum(row)
    items += len(row)
````

````python
total / items
````
