**(Discussion)** If you flip a coin a large odd number of times and associate the value 1 with heads and 2 with tails, what would you expect the mean to be? What would you expect the median and mode to be?

**Answer:** We'd expect the mean to be 1.5. As the number of flips increases, the head and tail percentages should each approach 50%. The median and mode will be whichever value has the higher frequency at a given time. If equal, the median would be 1.5, and the distribution would be bimodal with two modes—1 and 2.
