# Vibe Coding and Agentic Engineering with the Codex App 
### Vibe Coding
* Coined by Andrej Karpathy (2025)
    * AI researcher and OpenAI co-founder 
* Prompting AI tools to generate code
    > https://en.wikipedia.org/wiki/Vibe_coding  
    > https://www.ibm.com/think/topics/agentic-engineering

### Agentic Engineering
* Also coined by Karpathy
    > **"agentic"** because the new default is that you are not writing the code directly 99% of the time, you are orchestrating agents who do and acting as oversight.
    > 
    > **"engineering"** to emphasize that there is an art & science and expertise to it. It's something you can learn and become better at, with its own depth of a different kind."

    > * https://x.com/karpathy/status/2019137879310836075

### Where We're Headed — Harness Engineering: Agent-First Software Development
* **Core idea:** Humans steer, Codex agents execute
* **Engineers' role:** specify intent, design scaffolding, create feedback loops, and make agent work reliable
* OpenAI built a product with **0 manually written code**, about **one million lines**, **1500 pull requests**, and roughly **1/10th expected build time**
* Read the **OpenAI blog post:** https://openai.com/index/harness-engineering/
    * Term appears to have been coined by Viv Tevedi of LangChain (Feb 2026): https://x.com/Vtrivedy10/status/2031408954517971368
    * Some say Mitchell Hashimoto popularized the term in this article (March 2026): https://mitchellh.com/writing/my-ai-adoption-journey
