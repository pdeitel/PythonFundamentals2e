# Tools Overview
* A tool extends what the model can do beyond text generation
* Hosted tools provide OpenAI-managed capabilities
    * Work only with OpenAI models
* Custom function tools wrap your Python functions 
* MCP tools expose external tool servers via a standard protocol
* Agent loop decides when to call a tool, executes it, then sends the result back to model

Reference docs:
* https://openai.github.io/openai-agents-python/tools/ 

---


## OpenAI Hosted Tools
* `WebSearchTool` — Search the web
* `FileSearchTool` — Retrieve info from your OpenAI vector stores
* `CodeInterpreterTool` — Execute code in a sandboxed environment
* `HostedMCPTool` — Expose a Model Context Protocol (MCP) server's tools to the model
* `ImageGenerationTool` — Generate images 
* `ToolSearchTool` — Dynamically load tools on demand
---

## OpenAI Local Tools
* `ComputerTool` — Computer/browser automation
* `ShellTool` — Local execution and hosted container execution
* `ApplyPatchTool` — CRUD operations on codebase
---

## Custom Tools and Non-OpenAI Tools
* Functions as tools
* Agents as tools
* Tools accessed via Model Context Protocol (MCP)
    * Can be local or remove
