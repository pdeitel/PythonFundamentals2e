**(IPython Session)** Use `partition` and `rpartition` to extract from the URL string
```
'https://www.deitel.com/books/PyCDS/table_of_contents.html'
```
the substrings `'www.deitel.com'` and `'books/PyCDS'`.

````python
url = 'https://www.deitel.com/books/PyCDS/table_of_contents.html'
````

````python
protocol, separator, rest_of_url = url.partition('://')
````

````python
host, separator, document_with_path = rest_of_url.partition('/')
````

````python
path, separator, document = document_with_path.rpartition('/')
````
