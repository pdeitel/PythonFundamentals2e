### Checkpoint 2 Snippets

**(IPython Session)** Assume you have a string representing an addition problem, such as `'10 + 5'`. Use a regular expression to extract the substrings for the two operands and the operator, then display them using the Match object's `groups` and `group` methods.

```python
import re
```

```python
result = re.search(r'(\d+) ([-+*/]) (\d+)', '10 + 5')
```

```python
result.groups()
```

```python
result.group(1)
```

```python
result.group(2)
```

```python
result.group(3)
```
